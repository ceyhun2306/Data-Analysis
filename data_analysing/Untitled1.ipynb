#What is the distribution of Employee Status (Active, Resigned, Retired, Terminated) ?

```python
status_count = df['Status'].value_counts()

bars = plt.bar(status_count.index, status_count.values)
plt.xlabel("Employee Status")
plt.ylabel("Count")
plt.title("Status Distribution")
plt.bar_label(bars, status_count.values)
plt.show()
```

#What is the distribution of work modes (On-site, Remote) ?

```python
modes_counts = df['Work_Mode'].value_counts()

plt.pie(modes_counts, labels=modes_counts.index, autopct='%1.1f%%')
plt.title("Work Mode Distribution")
plt.show()
```

#How many employees are there in each department ?

```python
department_worker_count = df['Department'].value_counts()

bars = plt.bar(department_worker_count.index, department_worker_count.values, color="green")
plt.xlabel("Department names")
plt.xticks(rotation=30)
plt.ylabel("Employee Count")
plt.title("Employee Count for each Department")
plt.bar_label(bars, status_count.values)
plt.show()
```

#What is the average salary by Department ?

```python
avg_salary_dept = df.groupby("Department")["Salary_INR"].mean()

bars = plt.bar(avg_salary_dept.index, avg_salary_dept.values, color="orange")
plt.ylabel("Average Salary (INR)")
plt.xticks(rotation=15)
plt.title("Average Salary by Department")

plt.bar_label(bars, labels=[f"{v:.0f}" for v in avg_salary_dept.values])

plt.show()
```

#Which job title has the highest average salary ?

```python
avg_salary_jobTitle = df.groupby("Job_Title")["Salary_INR"].mean()

plt.figure(figsize=(24, 6))
bars = plt.bar(avg_salary_jobTitle.index, avg_salary_jobTitle.values, color=["lightBlue", "lightGreen", "yellow"])
plt.ylabel("Average Salary (INR)")
plt.xticks(rotation=45)
plt.title("Average Salary by Job title")

plt.bar_label(bars, labels=[f"{v:.0f}" for v in avg_salary_jobTitle.values])

plt.show()
```

```python
#Highest average salary in job title is IT Manager
df.groupby("Job_Title")["Salary_INR"].mean().max()
```

#What is the average salary in different Departments based on Job Title ?

```python
# Compute average salaries
avg_salary = df.groupby(["Job_Title", "Department"])["Salary_INR"].mean().unstack()

# Plot grouped bars
ax = avg_salary.plot(kind="bar", figsize=(24,6))
plt.ylabel("Average Salary (INR)")
plt.title("Average Salary by Job Title and Department")
plt.xticks(rotation=30)
plt.legend(title="Department")
plt.show()
```

#How many employees Resigned & Terminated in each department ?

```python
df_resigned = df[df['Status'].isin(['Resigned'])].copy()
count_resigned_per_dept = df_resigned.groupby('Department')['Status'].count()
print(count_resigned_per_dept)
```

```python
df_terminated = df[df['Status'].isin(['Terminated'])].copy()
count_terminated_per_dept = df_terminated.groupby('Department')['Status'].count()
print(count_terminated_per_dept)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14,6))

axes[0].bar(count_resigned_per_dept.index, count_resigned_per_dept.values)
axes[0].set_xlabel("Departments")
axes[0].set_ylabel("Counts")
axes[0].set_title("Resigned Employees by Department")
axes[0].tick_params(axis='x', rotation=30)

axes[1].bar(count_terminated_per_dept.index, count_terminated_per_dept.values, color='orange')
axes[1].set_xlabel("Departments")
axes[1].set_title("Terminated Employees by Department")
axes[1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

```python
print("Total: ")
df_filtered = df[df['Status'].isin(['Resigned', 'Terminated'])].copy()
count_per_dept = df_filtered.groupby('Department')['Status'].count()
print(count_per_dept)
```

#How does salary vary with years of experience ?

```python
df_exper_salary_byTitle = df.groupby(by=["Job_Title", "Experience_Years"])["Salary_INR"].mean().reset_index()

job_title = df['Job_Title'].unique()
for title in job_title:
    df_title = df_exper_salary_byTitle[df_exper_salary_byTitle['Job_Title'] == title]
    salaries_full = df_title["Salary_INR"]

    base = (salaries_full // 100000) * 100000
    remainder = salaries_full % 100000

    experience = df_title["Experience_Years"]

    plt.figure(figsize=(12, 4))
    bars = plt.bar(experience, remainder)

    plt.xlabel("Years of Experience")
    plt.ylabel("Salary (INR)")
    plt.title(f"Salary Trend by Experience for {title}")

    plt.bar_label(bars, labels=[f"{int(r):,}" for r in remainder])

    y_ticks = plt.yticks()[0]
    y_labels = [f"{int(t + base.min()):,}" for t in y_ticks]
    plt.yticks(y_ticks, y_labels)

    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
```

#What is the average performance rating by department ?

```python
avg_performance_dept = df.groupby("Department")["Performance_Rating"].mean().map(lambda x: round(x, 4))

bars = plt.bar(avg_performance_dept.index, avg_performance_dept.values, color='skyblue')
plt.ylabel("Average Performance Rating")
plt.title("Average Performance Rating by Department")

plt.xticks(rotation=30)

plt.bar_label(bars, labels=[f"{v:.4f}" for v in avg_performance_dept.values])

plt.ylim(min(avg_performance_dept.values) - 0.002, max(avg_performance_dept.values) + 0.002)

plt.show()
```

```python
plt.plot(avg_performance_dept.index, avg_performance_dept.values, marker='o')
plt.ylabel("Average Performance Rating")
plt.title("Average Performance Rating by Department")
plt.xticks(rotation=30)
plt.grid(True)
plt.show()
```

#Is there a correlation between performance rating and salary ?

```python
import seaborn as sns

corr_matrix = df.select_dtypes(include=np.number).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, square=True)
plt.title("Correlation Matrix of Numeric Columns")
plt.show()
```

#How has the number of hires changed over time (per year) ?

```python
df["Hire_Date"] = pd.to_datetime(df["Hire_Date"])

df["Hire_year"] = df["Hire_Date"].dt.year
```

```python
hire_years = df['Hire_year'].value_counts()

plt.figure(figsize=(16, 6))
bars = plt.bar(hire_years.index, hire_years.values, color='skyblue')
plt.xlabel("Hire Years")
plt.ylabel("Number of Employees")
plt.title("Hired counts by Year")
plt.bar_label(bars)
plt.show()
```
