```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/restaurant-sales-data.zip")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.head(10)
```

```python
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y")
df["Day"] = df["Date"].dt.day
df["Month"] = df["Date"].dt.month
df["Year"] = df["Date"].dt.year
df
```

```python
df["Manager"] = df["Manager"].str.strip()
df["Manager"] = df["Manager"].str.replace(r"\s+", " ", regex=True)
df["Manager"] = df["Manager"].str.title()
```

```python
df["Manager"].count()
```

#How different is total quantity if products?

```python
df.groupby("Product")["Quantity"].sum().plot(kind="bar", figsize=(8, 5), color="skyblue", edgecolor="black")
plt.title("Total Quantity Sold by Product")
plt.xlabel("Product")
plt.ylabel("Total Quantity")
plt.xticks(rotation=45)
plt.show()
```

# ================= Conclusion =================
# â€¢ Beverages recorded the highest sales quantity.  
# â€¢ Burgers and Fries are also close to the top in terms of sales.  
# â€¢ Chicken Sandwiches and Sides & Other have significantly lower sales quantities.  
# ==============================================


# How many sales does managers did?

```python
df.groupby("Manager")["Quantity"].sum().sort_values().plot(kind="barh", figsize=(8,5), color="lightgreen", edgecolor="black")
plt.title("Total Sales by Manager")
plt.xlabel("Total Quantity")
plt.ylabel("Manager")
plt.show()
```

# ================= Conclusion =================
# â€¢ Joao Silva achieved the highest total sales.  
# â€¢ Tom Jackson is very close behind him.  
# â€¢ The lowest sales were recorded by Remy Monet.  
# â€¢ Walter Muller performed only slightly better,
#   with around 5,000 more sales than Remy Monet.  
# ==============================================


#How much big is difference between payment methods?

```python
df["Payment Method"].value_counts().plot(kind="pie", autopct="%1.1f%%", startangle=90, figsize=(6,6), colors=["gold", "lightcoral", "skyblue"])
plt.title("Share of Payment Methods")
plt.ylabel("")
plt.xlabel("")
plt.show()
```

# ================= Conclusion =================
# â€¢ There is not much difference between Gift Card and Cash payments.  
# â€¢ However, Credit Card payments are significantly higher.  
# â€¢ Almost half of all customers prefer to pay by Credit Card.  
# ==============================================


#How much difference have between daily sales?

```python
df.groupby("Date")["Quantity"].sum().plot(kind="line", marker="o", figsize=(8, 5), color="purple")
plt.title("Daily Total Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Quantity")
plt.grid(True, linestyle="--")
plt.show()
```

# ================= Conclusion =================
# â€¢ Overall, there is not a major difference in sales between days.  
# â€¢ In general, sales have been increasing steadily since the start.  
# â€¢ However, a sharp decline was observed on December 20
#   and again on January 4.  
# ==============================================


#In each city how much different count of sales products(all 5 category of products)

```python
(df.pivot_table(index="City", columns="Product", values="Quantity", aggfunc="sum")
   .plot(kind="bar", figsize=(10,6)))
plt.title("Sales by City and Product")
plt.xlabel("City")
plt.ylabel("Total Quantity")
plt.xticks(rotation=45)
plt.legend(title="Product", bbox_to_anchor=(1.05,1))
plt.tight_layout()
plt.show()
```

# ================= Conclusion =================
# â€¢ Overall, sales were highest in Lisbon and London.  
# â€¢ Other cities had relatively lower sales.  
# â€¢ When comparing individual products, Beverages and Fries consistently
#   show their dominance in each city.  
# ==============================================


#How much is prices of products(Average)?

```python
df.groupby("Product")["Price"].mean().plot(kind="bar", figsize=(8,5), color="orange", edgecolor="black")
plt.title("Average Price by Product")
plt.xlabel("Product")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```

# ================= Conclusion =================
# â€¢ As observed, the average price of Burgers across all cities exceeds Â£12.  
# â€¢ The second most expensive product is Chicken Sandwiches, with an average price above Â£10.  
# â€¢ Beverages and Fries are the cheapest products, ranging between Â£2â€“Â£4.  
# â€¢ This price difference likely affects their sales quantities.  
# ==============================================


#Totaly how differetn are sales between cities?

#Which purchase type mostly people prefer?

```python
df.groupby("Purchase Type")["Quantity"].sum().plot(kind="barh", figsize=(8,5), color="teal", edgecolor="black")
plt.title("Total Sales by Purchase Type")
plt.xlabel("Total Quantity")
plt.ylabel("Purchase Type")
plt.show()
```

# ================= Conclusion =================
# â€¢ Customers prefer to make purchases online the most.  
# â€¢ Drive-thru purchases are the least preferred.  
# â€¢ In-store purchases fall in between these two methods.
